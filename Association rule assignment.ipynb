```python
'''1.	What is lift and why is it important in Association rules?

A . Lift is a measure of the strength of the association between two items in Association rules.
It is calculated as the ratio of the observed frequency of the items to the expected frequency if they were independent.
Lift is important because it helps to identify rules that are more likely to be meaningful and useful.'''
```

```python
'''
2.	What is support and Confidence. How do you calculate them?

A . Support shows how often something appears.
Confidence shows how often something happens when something else happens.

Support is calculated as:
Support(A) = Number of transactions containing A / Total number of transactions

Confidence is calculated as:
Confidence(A=>B) = Number of transactions containing both A and B / Number of transactions containing A
'''
```

```python
'''
3.	What are some limitations or challenges of Association rules mining?

A. Some challenges of Association rules mining are:

- Spurious rules: Rules that appear significant but are actually due to chance
- Redundant rules: Rules that are similar or identical to others
- Uninteresting rules: Rules that are obvious or not actionable
'''
```
